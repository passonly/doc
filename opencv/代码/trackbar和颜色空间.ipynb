### trackbar的使用

```python
import cv2
import numpy as np
```

```python
# createTrackbar(trackbarname, winname, 起始值, 最大值, callback)
# 通过trackbar修改背景颜色

# 创建窗口
cv2.namedWindow('trackbar', cv2.WINDOW_NORMAL)
cv2.resizeWindow('trackbar', 640, 800)

def callback(value):
    pass
# 创建trackbar
# 颜色通道有三种颜色, 红绿蓝. 
cv2.createTrackbar('R', 'trackbar', 0, 255, callback)
cv2.createTrackbar('G', 'trackbar', 0, 255, callback)
cv2.createTrackbar('B', 'trackbar', 0, 255, callback)

# 创建一个背景图片
img = np.zeros((480, 640, 3), np.uint8)

while True:
    # getTrackbarPos(trackbarname, winname) 获取trackbar的当前值
    r = cv2.getTrackbarPos('R', 'trackbar')
    g = cv2.getTrackbarPos('G', 'trackbar')
    b = cv2.getTrackbarPos('B', 'trackbar')
    
    # 用获取到的三个值去修改背景图片颜色
    img[:] = [b, g, r]
    cv2.imshow('trackbar', img)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
        
cv2.destroyAllWindows()
```

### 颜色空间

#### BGR和RGB

```python
# opencv默认读进来的图片是BGR颜色空间. 
cat = cv2.imread('./cat.jpeg')
cv2.imshow('cat', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 如果一个BGR颜色空间的图片, 放到RGB上显示, 颜色就会出问题
import matplotlib.pyplot as plt
# plt就是按照RGB来显示的. 
plt.imshow(cat)
```

```python
# 如果想显示正确, 可以把cat转化成正常RGB颜色空间
cat_rgb = cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
plt.imshow(cat_rgb)
```

#### HSV

```python
# opencv默认读进来的图片是BGR颜色空间. 
cat = cv2.imread('./cat.jpeg')
# 修改颜色空间 convert
cat_hsv = cv2.cvtColor(cat, cv2.COLOR_BGR2YUV)
# cv2.imshow是按照BGR来显示图片的. 
cv2.imshow('cat', cat_hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 颜色空间应用

```python
# 通过trackbar展示不同颜色空间的效果
# 创建窗口
import cv2
cv2.namedWindow('color', cv2.WINDOW_NORMAL)
cv2.resizeWindow('color', 640, 800)


def callback(value):
    pass

img = cv2.imread('./cat.jpeg')

# 定义颜色空间转化列表
color_spaces = [
    cv2.COLOR_BGR2GRAY, 
    cv2.COLOR_BGR2BGRA,
    cv2.COLOR_BGR2RGB,
    cv2.COLOR_BGR2RGBA,
    cv2.COLOR_BGR2HSV,
    cv2.COLOR_BGR2HLS,
    cv2.COLOR_BGR2YUV
]

# 设置trackbar
cv2.createTrackbar('trackbar', 'color', 0, 6, callback)

while True:
    # 获取trackbar的值
    index = cv2.getTrackbarPos('trackbar', 'color')
    # 进行颜色空间的转化
    cvt_img = cv2.cvtColor(img, color_spaces[index])
    cv2.imshow('color', cvt_img)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
# mat的深浅拷贝. 在opencv-python中对应着ndarray的深浅拷贝. 
# ndarray的深浅拷贝?
# 浅拷贝, view
# 深拷贝, copy
```

```python
import cv2
import numpy as np

img = cv2.imread('./cat.jpeg')

# 浅拷贝
img2 = img.view()

# 深拷贝
img3 = img.copy()

img[10: 100, 10:100] = [0, 0, 255]

cv2.imshow('img', np.hstack((img, img2, img3)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
