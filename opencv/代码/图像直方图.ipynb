```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 使用opencv统计直方图

```python
img = cv2.imread('./lena.png')
hist = cv2.calcHist([img], [0], None, [256], [0, 255])
# print(hist)
print(type(hist))
print(hist.shape)
# _ = plt.hist(hist.reshape(-1), bins=256, range=[0, 255])

plt.bar(np.arange(0, 256), hist.reshape(-1), width=1)
```

```python
# 使用灰度图来画直方图
img = cv2.imread('./lena.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gray], [0], None, [256], [0, 255])

_ = plt.hist(gray.ravel(), bins=256, range=[0, 255])
plt.show()

plt.bar(np.arange(0, 256), hist.reshape(-1), width=1)
```

```python
# 说明不能用颜色通道的直方图来代替灰度图的直方图. 甚至每个颜色通道的直方图都不一样. 
```

```python
img = cv2.imread('./lena.png')

histb = cv2.calcHist([img], [0], None, [256], [0, 255])
histg = cv2.calcHist([img], [1], None, [256], [0, 255])
histr = cv2.calcHist([img], [2], None, [256], [0, 255])

plt.plot(histb, color='b', label='blue')
plt.plot(histg, color='g', label='green')
plt.plot(histr, color='r', label='red')
plt.legend()
```

### 带掩膜(码)的直方图

```python
img = cv2.imread('./lena.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 生成掩膜
mask = np.zeros(gray.shape, np.uint8)

# 指定需要统计直方图的位置
mask[200:400, 200: 400] = 255

hist_mask = cv2.calcHist([gray], [0], mask, [256], [0, 255])
hist_gray = cv2.calcHist([gray], [0], None, [256], [0, 255])

cv2.imshow('gray', gray)
cv2.imshow('gray_mask', cv2.bitwise_and(gray, gray, mask=mask))


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.plot(hist_mask, label='mask')
plt.plot(hist_gray, label='gray')

plt.legend()
```

### 直方图均衡化

```python
gray = cv2.imread('./1.jpeg', 0)

hist = cv2.calcHist([gray], [0], None, [256], [0, 255])

plt.plot(hist)
plt.show()
plt.plot(hist_gray)
```

```python
gray_equ = cv2.equalizeHist(gray)

cv2.imshow('gray', gray)
cv2.imshow('gray_equ', gray_equ)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist_gray = cv2.calcHist([gray_equ], [0], None, [256], [0, 255])

plt.plot(hist_gray)
```

```python
gray = cv2.imread('./2.jpeg', 0)

hist = cv2.calcHist([gray], [0], None, [256], [0, 255])
gray_equ = cv2.equalizeHist(gray)
hist_gray = cv2.calcHist([gray_equ], [0], None, [256], [0, 255])

plt.plot(hist)
plt.show()
plt.plot(hist_gray)
```
