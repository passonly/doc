### 绘制轮廓

```python
img = cv2.imread('./contours1.jpeg')

# 一般查找轮廓都是使用黑白图. 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 二值化
thresh, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

# cv2.CHAIN_APPROX_SIMPLE只保存角点, cv2.CHAIN_APPROX_NONE保存所有的点. 
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# 绘制轮廓
img2 = cv2.drawContours(img.copy(), contours, -1, (0, 0, 255), 2)

cv2.imshow('img', img)
cv2.imshow('img2', img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 计算轮廓面积

```python
img = cv2.imread('./a2.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 二值化
binary = cv2.adaptiveThreshold(gray,  255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 9, 0)

contours, _ = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

img_copy = img.copy()
contours2 = []
# 计算轮廓面积
for i in range(len(contours)):
    area = cv2.contourArea(contours[i])
    print(area)
    if area > 100:
        contours2.append(contours[i])



cv2.drawContours(img_copy, contours2, -1, (0, 0, 255), 2)


cv2.imshow('binary', binary)
cv2.imshow('img', img)
cv2.imshow('img_copy', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 多边形逼近和凸包

```python
img = cv2.imread('./hand.png')

# 一般查找轮廓都是使用黑白图. 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 二值化
thresh, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(contours[0].shape)

# 使用多变形逼近, 减少存储的轮廓点. 
approx = cv2.approxPolyDP(contours[0], 5, closed=True)
print(approx.shape)
img_copy = img.copy()
cv2.drawContours(img_copy, [approx], 0, (0, 0, 255), 2)

cv2.imshow('img', img)
cv2.imshow('img_copy', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 凸包
img = cv2.imread('./hand.png')

# 一般查找轮廓都是使用黑白图. 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 二值化
thresh, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(contours[0].shape)

# 使用多变形逼近, 减少存储的轮廓点. 
approx = cv2.approxPolyDP(contours[0], 5, closed=True)
print(approx.shape)

# 凸包
hull = cv2.convexHull(contours[0])

img_copy = img.copy()
cv2.drawContours(img_copy, [hull], 0, (0, 0, 255), 2)

cv2.imshow('img', img)
cv2.imshow('img_copy', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 最小外矩形和最大(up-right)外接矩形

```python
img = cv2.imread('./hello.jpeg')
# 一般查找轮廓都是使用黑白图. 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 二值化
thresh, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)


# 最小外接矩形
# 返回的是旋转矩形 Rotated Rect, 包含矩形的起始坐标(x,y), 长宽, 旋转角度. 
rect = cv2.minAreaRect(contours[1])
print(rect)
# boxPoints可以帮助我们把旋转矩形的4个坐标计算出来. 
box = cv2.boxPoints(rect)
print(box)

# 变成整数
box = np.round(box).astype('int64')

# 最大外接矩形
# 返回矩形的x,y, w,h
x, y, w, h = cv2.boundingRect(contours[1])

img_copy = img.copy()
cv2.drawContours(img_copy, [box], 0, (0, 0, 255), 2)

cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)


cv2.imshow('img', img)
cv2.imshow('img_copy', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
