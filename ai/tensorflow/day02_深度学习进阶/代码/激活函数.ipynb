```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# relu max(0, x)
def relu(x):
    return max(0, x)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [relu(i) for i in x])
```

```python
# sigmoid一般作为分类任务最后一层(输出层)的激活函数, 一般不做隐藏层的激活函数. 因为不容易训练, 因为容易梯度消失或梯度爆炸. 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [sigmoid(i) for i in x])
```

```python
# tanh 双曲正切函数
# tanh也会有梯度消失和梯度爆炸的问题, 所以也不常用, 只在特定网络中使用. 比如RNN. 
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [tanh(i) for i in x])
```

```python
# softplus
def softplus(x):
    return np.log(np.exp(x) + 1)
```

```python
softplus(10)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, [softplus(i) for i in x])
plt.plot(x, [relu(i) for i in x])
```
