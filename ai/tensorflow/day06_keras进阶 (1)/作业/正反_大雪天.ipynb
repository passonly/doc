```python
import numpy as np
np.set_printoptions(suppress=True)
```

```python
def linear_z(x,w,b):
    return x@w + b
def sigmoid_a(z):
    return 1/(1+np.exp(-z))
```

```python
def dz_a(w,dz):
    return  dz@w.T
def da_z(a,da):
    return da*(a*(1-a))
```

```python
def dz_w(dz,a):
    return (a.T @ dz)/20
def dz_b(dz):
    return dz.mean(axis =0)
```

```python
X = np.random.randint(0,120,(20,12))
y = np.random.randint(0,2,(20,1))
W1 = np.random.randn(12,6)
B1 = np.zeros((1,6))
W2 = np.random.randn(6,4)
B2= np.zeros((1,4))
W3 = np.random.randn(4,2)
B3= np.zeros((1,2))
W4 = np.random.randn(2,1)
B4= np.zeros((1,1))
n = 3001
learning = 0.15*n/(n+1)

X_ = (X-X.mean(axis = 0))/X.std(axis=0)

while n:

    z1 = linear_z(X_,W1,B1)
    a1 = sigmoid_a(z1)
    z2 = linear_z(a1,W2,B2)
    a2 = sigmoid_a(z2)
    z3 = linear_z(a2,W3,B3)
    a3 = sigmoid_a(z3)
    z4 = linear_z(a3,W4,B4)
    a4 = sigmoid_a(z4)
 
    j = -y*(a4) - (1-y)*(1-a4)

    dz4 = (a4-y)
    da3 = dz_a(W4,dz4)
    dz3 = da_z(a3,da3)
    da2 = dz_a(W3,dz3)
    dz2 = da_z(a2,da2)
    da1 = dz_a(W2,dz2)
    dz1 = da_z(a1,da1)
 
    dw4 = dz_w(dz4,a3)
    dw3 = dz_w(dz3,a2)
    dw2 = dz_w(dz2,a1)
    dw1 = dz_w(dz1,X)

    db4 = dz_b(dz4)
    db3 = dz_b(dz3)
    db2 = dz_b(dz2)
    db1 = dz_b(dz1)
 

    W1 = W1 - learning * dw1
    B1 = B1 - learning * db1.T
    W2 = W2 - learning * dw2
    B2 = B2 - learning * db2.T
    W3 = W3 - learning * dw3
    B3 = B3 - learning * db3.T
    W4 = W4 - learning * dw4
    B4 = B4 - learning * db4.T
    
    n-=1
    if (3000-n)%150 == 0:
        print(3000-n)
        print(np.round(a4.T,3))
        print(y.T)
        print(j.sum())
        print('_____________')
        
```
