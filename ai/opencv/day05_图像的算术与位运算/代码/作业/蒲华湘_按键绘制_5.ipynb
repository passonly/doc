```python
#导包
import cv2
import numpy as np
```

```python
#函数
class line:
    def __init__(self):
        self.start_x = 0
        self.start_y = 0
        self.end_x = 0
        self.end_y = 0
        self.drawing = False
        self.move = False
        self.r = 0
        self.g = 0
        self.b = 0
        print('开始画线')
        
        
    def draw_start(self, event, x, y, flags, userdata):
        if event == cv2.EVENT_LBUTTONDOWN: # 按下左键 开始画
            self.drawing = True
            self.start_x,self.start_y = x, y  # 线的起点 就是 鼠标的点
            
        elif event == cv2.EVENT_MOUSEMOVE:# 鼠标移动
            self.img2 = self.img.copy()
            self.move = True
            if self.drawing:
                cv2.line(self.img2, (self.start_x,self.start_y), (x, y), (self.b, self.g, self.r),1, 16)
                
        elif event == cv2.EVENT_LBUTTONUP:  # 结束画画
            self.drawing = False
            self.move = False
            cv2.line(self.img, (self.start_x,self.start_y), (x, y), (self.b, self.g, self.r), 1, 16)

            
    def draw(self,name,img):
        #建立trackbar
        cv2.namedWindow('trackbar',cv2.WINDOW_NORMAL)
        def callback(value):
            pass
            # 创建trackbar
            # 颜色通道有三种颜色, 红绿蓝. 
        cv2.createTrackbar('R', 'trackbar', 0, 255, callback)
        cv2.createTrackbar('G', 'trackbar', 0, 255, callback)
        cv2.createTrackbar('B', 'trackbar', 0, 255, callback)
        
        self.img = img
        cv2.setMouseCallback(name,self.draw_start,'线')
        
        while True:
            #获取trackbar的值
            r = cv2.getTrackbarPos('R', 'trackbar')
            g = cv2.getTrackbarPos('G', 'trackbar')
            b = cv2.getTrackbarPos('B', 'trackbar')
            self.r = r
            self.g = g 
            self.b = b
            if self.move:
                cv2.imshow('window',self.img2)
            else:
                cv2.imshow('window',self.img)
            key = cv2.waitKey(1) 
            if key == ord('q'):
                print('退出划线')
                break
        cv2.destroyWindow('trackbar')
        
        


class rectangle:
    def __init__(self):
        self.start_x = 0
        self.start_y = 0
        self.end_x = 0
        self.end_y = 0
        self.drawing = False
        self.move = False
        print('开始画矩形')


    def draw_start(self, event, x, y, flags, userdata):
        if event == cv2.EVENT_LBUTTONDOWN: # 按下左键 开始画
            self.drawing = True
            self.start_x,self.start_y = x, y  # 矩形对角线的起点 就是 鼠标的点

        elif event == cv2.EVENT_MOUSEMOVE:# 鼠标移动
            self.img2 = self.img.copy()
            self.move = True
            if self.drawing:
                cv2.rectangle(self.img2, (self.start_x,self.start_y), (x, y), (self.b, self.g, self.r),1, 16)    

        elif event == cv2.EVENT_LBUTTONUP:  # 结束画画
            self.drawing = False
            cv2.rectangle(self.img, (self.start_x,self.start_y), (x, y), (self.b, self.g, self.r), 1, 16)


    def draw(self,name,img):
        #建立trackbar
        cv2.namedWindow('trackbar',cv2.WINDOW_NORMAL)
        def callback(value):
            pass
            # 创建trackbar
            # 颜色通道有三种颜色, 红绿蓝. 
        cv2.createTrackbar('R', 'trackbar', 0, 255, callback)
        cv2.createTrackbar('G', 'trackbar', 0, 255, callback)
        cv2.createTrackbar('B', 'trackbar', 0, 255, callback)
        self.img = img
        cv2.setMouseCallback(name,self.draw_start,'矩形')

        while True:
            #获取trackbar的值
            r = cv2.getTrackbarPos('R', 'trackbar')
            g = cv2.getTrackbarPos('G', 'trackbar')
            b = cv2.getTrackbarPos('B', 'trackbar')
            self.r = r
            self.g = g 
            self.b = b
            if self.move:
                cv2.imshow('window',self.img2)
            else:
                cv2.imshow('window',self.img)
            key = cv2.waitKey(1) 
            if key == ord('q'):
                print('退出矩形')
                break
        cv2.destroyWindow('trackbar')

class circle:
    def __init__(self):
        self.start_x = 0
        self.start_y = 0
        self.end_x = 0
        self.end_y = 0
        self.drawing = False
        self.move = False
        print('开始画圆')


    def draw_start(self, event, x, y, flags, userdata):
        if event == cv2.EVENT_LBUTTONDOWN: # 按下左键 开始画
            self.drawing = True
            self.start_x,self.start_y = x, y  # 圆的起点 就是 鼠标的点
        elif event == cv2.EVENT_MOUSEMOVE:# 鼠标移动
            self.img2 = self.img.copy()
            s_x = pow(self.start_x-x,2) 
            s_y = pow(self.start_ｙ-ｙ,2)
            s = s_x + s_y
            r = pow(s,0.5)
            self.move = True
            if self.drawing:
                cv2.circle(self.img2, (self.start_x,self.start_y), int(r), (self.b, self.g, self.r),1, 16)
        elif event == cv2.EVENT_LBUTTONUP:  # 结束画画
            self.drawing = False
            s_x = pow(self.start_x-x,2) 
            s_y = pow(self.start_ｙ-ｙ,2)
            s = s_x + s_y
            r = pow(s,0.5)
            cv2.circle(self.img, (self.start_x,self.start_y), int(r), (self.b, self.g, self.r), 1, 16)


    def draw(self,name,img):
        #建立trackbar
        cv2.namedWindow('trackbar',cv2.WINDOW_NORMAL)
        def callback(value):
            pass
            # 创建trackbar
            # 颜色通道有三种颜色, 红绿蓝. 
        cv2.createTrackbar('R', 'trackbar', 0, 255, callback)
        cv2.createTrackbar('G', 'trackbar', 0, 255, callback)
        cv2.createTrackbar('B', 'trackbar', 0, 255, callback)
        self.img = img
        cv2.setMouseCallback(name,self.draw_start,'圆')

        while True:
            #获取trackbar的值
            r = cv2.getTrackbarPos('R', 'trackbar')
            g = cv2.getTrackbarPos('G', 'trackbar')
            b = cv2.getTrackbarPos('B', 'trackbar')
            self.r = r
            self.g = g 
            self.b = b
            
            if self.move:
                cv2.imshow('window',self.img2)
            else:
                cv2.imshow('window',self.img)
            key = cv2.waitKey(1) 
            if key == ord('q'):
                print('退出圆')
                break
                
        cv2.destroyWindow('trackbar')
    
```

```python
#主体
#初始化
cv2.namedWindow('window',cv2.WINDOW_NORMAL)
img = np.zeros((640, 640, 3), np.uint8)
cv2.imshow('window',img)
print('开始后,\nl画直线\nj画矩形\nr画圆\n每次进入模式后按q,选择模式再按q退出')


while True:
    key = cv2.waitKey(0)
    #三个模式
    if key == ord('l'):
        run = line()
        run.draw('window',img)
        
    elif key == ord('j'):
        run = rectangle()
        run.draw('window',img)
    
    elif key == ord('r'):
        run = circle()
        run.draw('window',img)
    
    elif key == ord('q'):
        print('结束')
        break
        
#关闭 
cv2.destroyAllWindows()
```
