```python
import cv2
import numpy as np
```

```python
# sheng《opencv作业-鼠标绘制图形》v1：初始版本
img = np.zeros((600,600,3),np.uint8)

def mouse_callback(event, x, y, flags, userdata):
    # 按下左键，获取起始坐标
    if event == cv2.EVENT_LBUTTONDOWN:
        global pt1 # 这里要设置全局
        pt1 = (x,y)
    
    # 松开左键，获取结束坐标
    elif event == cv2.EVENT_LBUTTONUP:
        pt2 = (x,y)
        if userdata == ord('l'):
            cv2.line(img,pt1,pt2,[0,0,255],2,16)
            cv2.imshow('img',img)
        
        elif userdata == ord('r'):
            cv2.rectangle(img,pt1,pt2,[0,0,255],2,16)
            cv2.imshow('img',img)
        
        elif userdata == ord('c'):
            r = np.sqrt((pt1[0]-pt2[0])**2 + (pt1[1]-pt2[1])**2 )
            cv2.circle(img,pt1,int(r),[0,0,255],2,16)
            cv2.imshow('img',img)

while True:
    cv2.imshow('img',img)
    key = cv2.waitKey(0) 
    if key == ord('q'):
        cv2.destroyAllWindows()
        break
    cv2.setMouseCallback('img',mouse_callback,key)

cv2.destroyAllWindows()
```

```python
# sheng《opencv作业-鼠标绘制图形》v2：显示绘图轨迹，文本提示
import cv2
import numpy as np

img_zero = np.zeros((600,600,3), np.uint8)
img = img_zero.copy()

# 文本提示
cv2.putText(img, "Key Tips:",(10,530), cv2.FONT_HERSHEY_SIMPLEX, 0.5, [0, 0, 255])
cv2.putText(img, "Press 'L' to draw line",(10,545), cv2.FONT_HERSHEY_SIMPLEX, 0.5, [0, 0, 255])
cv2.putText(img, "Press 'R' to draw rectangle",(10,560), cv2.FONT_HERSHEY_SIMPLEX, 0.5, [0, 0, 255])
cv2.putText(img, "Press 'C' to draw circle",(10,575), cv2.FONT_HERSHEY_SIMPLEX, 0.5, [0, 0, 255])
cv2.putText(img, "Press 'Q' to exit",(10,590), cv2.FONT_HERSHEY_SIMPLEX, 0.5, [0, 0, 255])

def mouse_callback(event, x, y, flags, userdata):
    global img_temp,pt0,pt1 # 声明全局，方便修改
    
    # 按下左键，获取起始坐标
    if event == cv2.EVENT_LBUTTONDOWN:
        pt0 = (x,y) # 起始点
        pt1 = (x,y) # 中间点
   
    # 初始化img_temp
    if not np.unique(img_temp).sum():
        img_temp = img_zero.copy()
            
    # 绘制鼠标拖动轨迹
    if userdata == ord('l'):
        if flags == 1 and event == cv2.EVENT_MOUSEMOVE:
            pt2 = (x,y) # 轨迹的终点
            cv2.line(img_temp,pt1,pt2,[0,0,255],2,16) 
            cv2.imshow('img',img_temp)
            pt1 = (x,y) # 轨迹的起点

    elif userdata == ord('r'):
        if flags == 1 and event == cv2.EVENT_MOUSEMOVE:
            pt2 = (x,y)
            cv2.rectangle(img_temp,pt0,pt2,[0,0,255],2,16) 
            cv2.imshow('img',img_temp)

    elif userdata == ord('c'):
        if flags == 1 and event == cv2.EVENT_MOUSEMOVE:
            pt2 = (x,y)
            r = np.sqrt((pt0[0]-pt2[0])**2 + (pt0[1]-pt2[1])**2 )
            cv2.line(img_temp,pt1,pt2,[0,0,255],2,16) # 显示半径
            cv2.circle(img_temp,pt0,int(r),[0,0,255],2,16)
            cv2.imshow('img',img_temp)

    # 清除鼠标拖动轨迹
    if not np.unique(img).sum():
        img_temp = img_zero.copy()

    # 根据按键，绘制图形
    if event == cv2.EVENT_LBUTTONUP: # 松开左键，获取结束坐标
        pt2 = (x,y)
        if userdata == ord('l'):    
            cv2.imshow('img',img_temp) 
            cv2.line(img,pt0,pt2,[0,0,255],2,16)
            cv2.imshow('img',img)
            img_temp = img.copy() # 拷贝当前img作为下次绘制轨迹的画板
 
        elif userdata == ord('r'):
            cv2.imshow('img',img_temp)
            cv2.rectangle(img,pt0,pt2,[0,0,255],2,16)
            cv2.imshow('img',img)
            img_temp = img.copy()
        
        elif userdata == ord('c'):
            cv2.imshow('img',img_temp)
            r = np.sqrt((pt0[0]-pt2[0])**2 + (pt0[1]-pt2[1])**2 )
            cv2.circle(img,pt0,int(r),[0,0,255],2,16)
            cv2.imshow('img',img)
            img_temp = img.copy()
            
    return img_temp

while True:
    cv2.imshow('img',img)
    key = cv2.waitKey(0)
    if key == ord('q'):
        cv2.destroyAllWindows()
        break
    img_temp = cv2.setMouseCallback('img',mouse_callback,key)
    
cv2.destroyAllWindows()
```
